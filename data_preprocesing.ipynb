### DATA PREPROCESSING

```python
# Step 1: Clean raw text
import re

# read raw text
with open("data/meitei_corpus.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Cleaning rules
text = re.sub(r'[_]+', '', text)    # remove ______
text = re.sub(r'\d+', '', text)     # remove numbers
text = re.sub(r'\n+', '\n', text)   # collapse multiple blank lines
text = re.sub(r"[()a-zA-Z'\"?/;.,\-#:%]", '', text)

text = text.strip()

# save cleaned text
with open("data/cleaned.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("✅ Cleaning done. Output saved to cleaned.txt")
```

```python

import re

with open("data/cleaned.txt", "r", encoding="utf-8") as f:
    text = f.read()

# separate markers
text = re.sub(r'\|\|', ' || ', text)
text = re.sub(r'꯫', ' ꯫ ', text)

tokens = text.split()

conll_lines = []

for token in tokens:
    if token in ["||", "꯫"]:
        # sentence boundary → blank line
        conll_lines.append("")
    else:
        conll_lines.append(f"{token} O")

with open("data/conll_O.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(conll_lines))

print("✅  CoNLL file ready")
```
